```python
def plot_images(images, tile_shape):
    assert images.shape[0] <= (tile_shape[0]* tile_shape[1])
    from mpl_toolkits.axes_grid1 import ImageGrid
    fig = plt.figure()
    grid = ImageGrid(fig, 111,  nrows_ncols = tile_shape ) 
    for i in range(images.shape[0]):
        grd = grid[i]
        grd.imshow(images[i])
```

```python
f = np.array(net.params['conv1'][0].data.squeeze())
# f = np.rollaxis(f, 0, 4)
# f = net.params['conv1'][0].data
# print f
plot_images(np.array(f), tile_shape=(4,5))
```

```python
plot_images(np.array(net.blobs['conv1'].data[0]), tile_shape=(4,5))
```

```python
f = np.array(net.params['conv2'][0].data.squeeze())
f = np.array(np.sum(f, 1))
print f.shape
plot_images(np.array(f), tile_shape=(5, 10))
```

```python
f = np.array(net.params['ip1'][0].data.squeeze()).reshape(10, 10, 6, 6)
f = np.array(np.sum(f, 1))
plot_images(np.array(f), tile_shape=(2, 5))
```

```python
f = np.array(net.params['ip1'][0].data.squeeze()).reshape(10, 10, 6, 6)
f = np.array(np.sum(f, 0))
plot_images(np.array(f), tile_shape=(2, 5))
```

## 性能評価

```python
import leveldb
from caffe.proto import caffe_pb2
db = leveldb.LevelDB('data/lmdb/test/separated_test_leveldb/')
test_data_raw =  [v for v in db.RangeIter()]
def toDatum(x):
    return caffe_pb2.Datum.FromString(x[1])
test_datum_set = map(toDatum, test_data_raw)
test_data = np.rollaxis(np.array(map(caffe.io.datum_to_array, test_datum_set)),1,4)
test_label = np.array(map(lambda x: x.label, test_datum_set))
```

```python
x = list(test_label[0:1936])
input_data = [d for d in test_data[:1936] / 255.0]
predictions_raw = net.predict(input_data)
y = list(predictions_raw.transpose(1, 0)[0, :1936] * 255)
# print "caffemodelの出力"
# predictions_raw
```

```python
y_new = []
for y_predict in y:
    print y_predict
    if y_predict >= 0.625 * 255:
        y_predict = 0
    else:
        y_predict = 255
    y_new.append(y_predict)
```
